```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_colwidth',120)
data1 = pd.read_csv('sample_1000_1 (Reeshad).csv')
data2 = pd.read_csv('sample_1000_1(Anita).csv')
data3 = pd.read_csv('sample_1000_2(Divyansh).csv')
data4 = pd.read_csv('sample_1000_2(Kirtan).csv')
main_data = pd.read_csv('park_tweet_csv_file.csv')
```

```python
# normalizing data
i = 0
data1 = data1.fillna(5)
for Actionable in data1['Actionability']:
    if Actionable == 5:
        data1.at[data1.index[i],'Actionability'] = 'Not Assigned'
    elif Actionable == 'NA':
        data1.at[data1.index[i],'Actionability'] = 'F'
    elif Actionable == 'A':
        data1.at[data1.index[i],'Actionability'] = 'T'

        
    i = i + 1
```

```python
i = 0
data2 = data2.fillna(5)
for Actionable in data2['Actionability']:
    if Actionable == 5:
        data2.at[data2.index[i],'Actionability'] = 'Not Assigned'
    elif Actionable == 'N':
        data2.at[data2.index[i],'Actionability'] = 'F'
    elif Actionable == 'Y':
        data2.at[data2.index[i],'Actionability'] = 'T'
        
    i = i + 1
```

```python
i = 0
data3 = data3.fillna(5)
for Actionable in data3['Actionability']:
    if Actionable == 5:
        data3.at[data3.index[i],'Actionability'] = 'Not Assigned'
    elif Actionable == False:
        data3.at[data3.index[i],'Actionability'] = 'F'
    elif Actionable == True:
        data3.at[data3.index[i],'Actionability'] = 'T'
        
    i = i + 1
```

```python
i = 0
data4 = data4.fillna(5)
for Actionable in data4['Actionability']:
    if Actionable == 5:
        data4.at[data4.index[i],'Actionability'] = 'Not Assigned'
    elif Actionable == 'f':
        data4.at[data4.index[i],'Actionability'] = 'F'
    elif Actionable == 't':
        data4.at[data4.index[i],'Actionability'] = 'T'
        
    i = i + 1
```

```python
#combining the similar datasets, Reeshad and anita's datasets
i = 0
for Actionable in data2['Actionability']:
    if Actionable == 'Not Assigned':
        data1.at[data1.index[i],'Actionability_2'] = 'Not Assigned'
    elif Actionable == 'F':
        data1.at[data1.index[i],'Actionability_2'] = 'F'
    elif Actionable == 'T':
        data1.at[data1.index[i],'Actionability_2'] = 'T'

        
    i = i + 1

print(data1['Actionability'])
print(data1['Actionability_2'])
```

```python
#combining the similar datasets, Reeshad and anita's datasets
i = 0
for Actionable in data4['Actionability']:
    if Actionable == 'Not Assigned':
        data3.at[data3.index[i],'Actionability_2'] = 'Not Assigned'
    elif Actionable == 'F':
        data3.at[data3.index[i],'Actionability_2'] = 'F'
    elif Actionable == 'T':
        data3.at[data3.index[i],'Actionability_2'] = 'T'

        
    i = i + 1
```

```python
# adding emotions to the tweets
i = 0

for text in data1['text']:
    j = 0 
    for text1 in main_data['text']:
        if text == text1:
            data1.at[data1.index[i],'emotion'] = main_data.at[main_data.index[j],'emotion']
            data1.at[data1.index[i],'emotion_score'] = main_data.at[main_data.index[j],'emotion_score']
        else:
            j = j + 1
        
    i = i + 1
```

```python
i = 0

for text in data3['text']:
    j = 0 
    for text1 in main_data['text']:
        if text == text1:
            data3.at[data3.index[i],'emotion'] = main_data.at[main_data.index[j],'emotion']
            data3.at[data3.index[i],'emotion_score'] = main_data.at[main_data.index[j],'emotion_score']
        else:
            j = j + 1
        
    i = i + 1
    
```

```python
# dropping rows with NaN emotions
data1 = data1.fillna(5)
data1.drop(data1[(data1['emotion'] == 5 )].index, inplace=True)


data3 = data3.fillna(5)
data3.drop(data3[(data3['emotion'] == 5 )].index, inplace=True)


data = pd.concat([data1, data3], ignore_index=True)
```

```python
print(data)
```

```python
# counting emotions with date
data[['created_at_date']] = data[['created_at']].apply(pd.to_datetime) 
data['created_at_date'] = data['created_at_date'].dt.date
data = data.sort_values(by=['created_at_date'])

data_sorted = data.copy()
```

```python
# counting actionable and number of emotional tweets over the sorted database, 
# This counts starts from 0 and reaches the actual total when it reaches the last row, the data isn't distributed by time




count_fear = 0
count_anger = 0
count_joy = 0
count_love = 0
count_surprise = 0
count_sadness = 0
count_action = 0

i = 0

for emoti in data_sorted['emotion']:

    if emoti == 'fear':
        count_fear = count_fear + 1

    if emoti == 'anger':
        count_anger = count_anger + 1
        
    if emoti == 'joy':
        count_joy = count_joy + 1 
        
    if emoti == 'love':
        count_love = count_love + 1 
        
    if emoti == 'surprise':
        count_surprise = count_surprise + 1 
        
    if emoti == 'sadness':
        count_sadness = count_sadness + 1 
        
    if data_sorted.at[data_sorted.index[i],'Actionability'] == 'T' or data_sorted.at[data_sorted.index[i],'Actionability_2'] == 'T':
        count_action = count_action + 1
       
       
    data_sorted.at[data_sorted.index[i],'count_of_fear_emotion'] = count_fear
    data_sorted.at[data_sorted.index[i],'count_of_anger_emotion'] = count_anger
    data_sorted.at[data_sorted.index[i],'count_of_joy_emotion'] = count_joy
    data_sorted.at[data_sorted.index[i],'count_of_love_emotion'] = count_love
    data_sorted.at[data_sorted.index[i],'count_of_surprise_emotion'] = count_surprise
    data_sorted.at[data_sorted.index[i],'count_of_sadness_emotion'] = count_sadness
    data_sorted.at[data_sorted.index[i],'count_Actionability'] = count_action
        
    i= i +1 
    


total_count = count_fear + count_anger + count_joy + count_love + count_surprise + count_sadness

print('percentage of fearful tweets : ' + ((str)((count_fear/total_count) *100)  +'%'))
print('percentage of angry tweets : ' + ((str)((count_anger/total_count) *100)  +'%'))
print('percentage of joyful tweets : ' + ((str)((count_joy/total_count) *100)  +'%'))
print('percentage of lovable tweets : ' + ((str)((count_love/total_count) *100)  +'%'))
print('percentage of surprise tweets : ' + ((str)((count_surprise/total_count) *100)  +'%'))
print('percentage of sadness tweets : ' + ((str)((count_sadness/total_count) *100)  +'%'))


```
